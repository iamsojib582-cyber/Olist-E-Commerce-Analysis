```python
import re
import html
import unicodedata
from urllib.parse import unquote

def normalize_city(text):
    if pd.isna(text):
        return text
    
    text = str(text).strip().lower()
    
    text = html.unescape(text)
    text = unquote(text)
    
    text = re.sub(r'^[\.\*\-_,;:]+', '', text)
    
    text = re.sub(r"[^a-zA-ZÀ-ÿ\s]", " ", text)
    
    text = re.sub(r'\b\w*\d+\w*\b', ' ', text)
    
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

def remove_accents(text):
    if pd.isna(text):
        return text
    return ''.join(
        ch for ch in unicodedata.normalize('NFKD', text)
        if not unicodedata.combining(ch)
    )

def pt_title(text):
    if pd.isna(text):
        return text
    
    small_words = {'de', 'da', 'do', 'das', 'dos', 'e'}
    words = text.split()
    
    result = []
    for i, word in enumerate(words):
        if i > 0 and word in small_words:
            result.append(word)
        else:
            result.append(word.capitalize())
    return ' '.join(result)

geolocation_df['city_clean_temp'] = geolocation_df['geolocation_city'].apply(normalize_city)

geolocation_df['city_key'] = geolocation_df['city_clean_temp'].apply(remove_accents)

canonical_map = (
    geolocation_df.groupby('city_key')['city_clean_temp']
    .agg(lambda x: x.value_counts().index[0])
    .to_dict()
)

geolocation_df['geolocation_city_cleaned'] = geolocation_df['city_key'].map(canonical_map)

geolocation_df['geolocation_city_cleaned'] = geolocation_df['geolocation_city_cleaned'].apply(pt_title)


geolocation_df['geolocation_city'] = geolocation_df['geolocation_city_cleaned']

geolocation_df.drop(columns=['city_clean_temp', 'city_key', 'geolocation_city_cleaned'], inplace=True)

print("City names cleaned.")
```

```python
import pandas as pd
import re
import html
import unicodedata
from urllib.parse import unquote

customers_df = pd.read_csv("olist_customers_dataset.csv")

def normalize_city(text):
    if pd.isna(text):
        return text
    
    text = str(text).strip().lower()
    text = html.unescape(text)
    text = unquote(text)
    text = re.sub(r'^[\.\*\-_,;:]+', '', text)
    text = re.sub(r"[^a-zA-ZÀ-ÿ\s]", " ", text)
    text = re.sub(r'\b\w*\d+\w*\b', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def remove_accents(text):
    if pd.isna(text):
        return text
    return ''.join(
        ch for ch in unicodedata.normalize('NFKD', text)
        if not unicodedata.combining(ch)
    )

def pt_title(text):
    if pd.isna(text):
        return text
    
    small_words = {'de', 'da', 'do', 'das', 'dos', 'e'}
    words = text.split()
    
    result = []
    for i, word in enumerate(words):
        if i > 0 and word in small_words:
            result.append(word)
        else:
            result.append(word.capitalize())
    return ' '.join(result)

customers_df['city_clean_temp'] = customers_df['customer_city'].apply(normalize_city)
customers_df['city_key'] = customers_df['city_clean_temp'].apply(remove_accents)

canonical_map = (
    customers_df.groupby('city_key')['city_clean_temp']
    .agg(lambda x: x.value_counts().index[0])
    .to_dict()
)

customers_df['customer_city'] = customers_df['city_key'].map(canonical_map).apply(pt_title)

customers_df.drop(columns=['city_clean_temp', 'city_key'], inplace=True)

print("customer_city cleaned successfully.")
```

```python
import pandas as pd
import unicodedata

def remove_accents(text):
    text = str(text)
    return ''.join(
        ch for ch in unicodedata.normalize('NFKD', text)
        if not unicodedata.combining(ch)
    )
sellers_df['seller_city'] = sellers_df['seller_city'].str.lower().str.strip()

sellers_df['seller_city'] = sellers_df['seller_city'].apply(remove_accents)

sellers_df['seller_city'] = sellers_df['seller_city'].replace({
    'vendas@creditparts.com.br': 'indaiatuba',
    "santa barbara d´oeste": "santa barbara d'oeste",
    'rio de janeiro \\rio de janeiro': 'rio de janeiro'
})

sellers_df['seller_city'] = sellers_df['seller_city'].str.split('\\').str[0].str.split('/').str[0].str.strip()
```
